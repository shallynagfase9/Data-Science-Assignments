```python
# Q1. How can you create a Bokeh plot using Python code?
'''
To create a Bokeh plot using Python code, you can follow these steps:
1. Import the necessary functions from the Bokeh library.
2. Define the data you want to plot.
3. Create a figure object and customize it with titles, axis labels, etc.
4. Add renderers to the figure to create the desired plot (e.g., line chart, scatter plot, bar chart).
5. Display the plot using the show() function.
'''
from bokeh.plotting import figure, show
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p = figure(title="Simple Line Chart", x_axis_label='X-axis', y_axis_label='Y-axis')
p.line(x, y, line_width=2)

show(p)
```

```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
'''
Glyphs in Bokeh refer to the visual shapes used to represent data in plots, such as circles, squares, lines, rectangles, etc.

To add glyphs to a Bokeh plot, you typically follow these steps:
1. Create a figure object using the figure() function.
2. Use specific glyph methods like circle(), line(), rect(), etc., to add different types of glyphs to the plot.
3. Customize the glyphs by setting properties like size, color, alpha, etc.
4. Display the plot using the show() function.
'''

from bokeh.plotting import figure, show
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p = figure(title="Circle Glyph Example", x_axis_label='X-axis', y_axis_label='Y-axis')
p.circle(x=x, y=y, size=10, color='blue', alpha=0.5)

show(p)
```

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
'''
To customize the appearance of a Bokeh plot, including the axes, title, and legend, you can use various attributes and methods provided by the Bokeh library.
'''
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers
output_file('test.html')

p = figure(title = "test flower")
p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"
p.circle(flowers["petal_length"],flowers["petal_width"],legend_label= "this is my sample graph")
show(p)
```
