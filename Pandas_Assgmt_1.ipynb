```python
# Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.
"""
A DataFrame in Pandas is a two-dimensional labeled data structure resembling a spreadsheet or SQL table, consisting of columns of potentially different types.
On the other hand, a Pandas Series is a one-dimensional labeled array that holds data of any type, similar to a column in a table.
"""
df = pd.DataFrame({
    "Name":["shally","harpreet"],
    "Age":[20, 21]
})
print(df)

data = pd.Series(data=[1,2,3,4,5])
print(data)
```
