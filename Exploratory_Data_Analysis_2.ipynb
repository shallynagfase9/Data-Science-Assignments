```python
"""
Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different
app categories.
"""
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('googleplaystore.csv')  # Replace 'googleplaystore.csv' with the correct path to your dataset

# Drop rows with missing values in the 'Rating' and 'Category' columns
data.dropna(subset=['Rating', 'Category'], inplace=True)

# Create a boxplot to compare ratings by category
plt.figure(figsize=(12, 8))
sns.boxplot(x='Category', y='Rating', data=data)
plt.title('Rating of Apps by Category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
"""
Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize
the relationship.
"""
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('googleplaystore.csv')  # Replace 'googleplaystore.csv' with the correct path to your dataset

# Drop rows with missing values in the 'Size' and 'Rating' columns
data.dropna(subset=['Size', 'Rating'], inplace=True)

# Convert the 'Size' column to numeric (removing 'M' and 'k' characters)
data['Size'] = data['Size'].apply(lambda x: float(x[:-1]) * 1000 if 'k' in x else float(x[:-1]))

# Create a scatter plot to visualize the relationship between app size and rating
plt.figure(figsize=(10, 6))
plt.scatter(data['Size'], data['Rating'], alpha=0.5)
plt.title('Relationship between App Size and Rating')
plt.xlabel('Size (KB)')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

```python
"""
Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.
"""

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('googleplaystore.csv')  # Replace 'googleplaystore.csv' with the correct path to your dataset

# Drop rows with missing values in the 'Type' and 'Price' columns
data.dropna(subset=['Type', 'Price'], inplace=True)

# Convert the 'Price' column to numeric (removing '$' character)
data['Price'] = data['Price'].apply(lambda x: float(x[1:]) if '$' in x else 0)

# Calculate average price by app type
avg_price_by_type = data.groupby('Type')['Price'].mean()

# Create a bar chart to compare average prices by app type
plt.figure(figsize=(10, 6))
avg_price_by_type.plot(kind='bar', color='skyblue')
plt.title('Average Price by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
"""
Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs.

"""
import pandas as pd

# Load the dataset
data = pd.read_csv('googleplaystore.csv')  # Replace 'googleplaystore.csv' with the correct path to your dataset

# Drop rows with missing values in the 'Installs' column
data.dropna(subset=['Installs'], inplace=True)

# Function to convert installs to numeric
def convert_installs(installs):
    installs = installs.replace(',', '')
    installs = installs.replace('+', '')
    return int(installs)

# Convert the 'Installs' column to numeric
data['Installs'] = data['Installs'].apply(convert_installs)

# Group data by app and calculate the total number of installs for each app
install_counts = data.groupby('App')['Installs'].sum()

# Sort the frequency table in descending order and get the top 10 most popular apps
top_10_apps = install_counts.sort_values(ascending=False).head(10)

print("Top 10 Most Popular Apps (Based on Number of Installs):\n")
print(top_10_apps)
```
