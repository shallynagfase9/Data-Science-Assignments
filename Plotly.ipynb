```python
# Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.
import seaborn as sns
titanic = sns.load_dataset("titanic")
import plotly.express as px
fig = px.scatter(titanic, x="age", y ="fare", color="sex",title="Scatter Plot")
fig.show()
```

```python
# Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
import plotly.express as px
df = px.data.tips()
fig = px.box(df, x='sex', y='total_bill')
fig.show()
```

```python
# Q5. What is Distplot? Using Plotly express, plot a distplot.
'''
A distplot, also known as a distribution plot, is a graphical representation that combines a histogram with a kernel density estimate (KDE) or a normal curve,
along with a rug plot.
It provides a visual representation of the distribution of data.
'''
import plotly.figure_factory as ff
import numpy as np

x1 = np.random.randn(200) - 1
x2 = np.random.randn(200)
x3 = np.random.randn(200) + 1

hist_data = [x1, x2, x3]
group_labels = ['Group 1', 'Group 2', 'Group 3']


fig = ff.create_distplot(hist_data, group_labels, bin_size=0.5)
fig.show()
```
